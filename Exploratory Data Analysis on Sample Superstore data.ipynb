# Categors vs profit

```python
%matplotlib inline
plt.figure(figsize=(40,25))
sns.barplot(x=data['Sub-Category'], y=data['Profit'])
```

From above two graphs, it can be noticed that the company is facing loss in two products such as bookcases and tables where the losses is more . Also, to be on brighter side, the company have lots of profits in copier but the number of sales is very less so there is need to increase number of sales in copier.

```python
state = data.groupby("State")[["Sales","Profit"]].sum().sort_values(by = "Sales", ascending = False)
plt.figure(figsize = (30,30))
state[:25].plot(kind = "bar",color = ["k","yellow"],edgecolor = "#000000")
plt.title("Profit Or Loss & Sales of top 25 States")
plt.xlabel("States")
plt.ylabel("Total profit / loss and Sales")
plt.grid(True)
state[25:].plot(kind = "bar",color = ["k","yellow"],edgecolor = "#000000")
plt.title("Profit Or Loss & Sales of the least economic States")
plt.xlabel("States")
plt.ylabel("Total profit / loss and Sales")
plt.grid(True)
```

From above graph, it can be examined that in top 25 States the company is facing more amount of loss whereas in least economic states the profits and sales are higher with loss only in Oregon.

```python
data.groupby('Sub-Category')['Profit','Sales'].sum().plot.bar(color = ['b','k'])
```

From above graph, the sales is higher in chairs and phones whereas the profits is gained through copier, accessories, papers. Overall, Phones showed both sales and profits whereas bookcases and supplies showed reverse trend.

```python
pd.DataFrame(data.groupby('State').sum())['Profit'].sort_values(ascending=True)
```

```python
pd.DataFrame(data.groupby('State').sum())['Discount'].sort_values(ascending=True)
```

```python
plt.figure(figsize = (12,6))
sns.lineplot('Discount', 'Profit', data = data, color = 'r', label= 'Discount')
plt.legend()
```

From the above study, when the profits of a state are compared with the discount provided in each state, the states which allowed more discount, went in loss and the above graph we conclude that THE PROFIT DECREASES WITH THE INCREASE IN DISCOUNT.
